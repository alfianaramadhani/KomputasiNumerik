## Menggunakan model yang ada di PPT

$$
f(x) = a \ln(x) + b \cos(x) + c e^x
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Data dari slide
x = np.array([0.24, 0.65, 0.95, 1.24, 1.73, 2.01, 2.23, 2.52])
y = np.array([0.23, -0.23, -1.1, -0.45, 0.27, 0.1, -0.29, 0.24])

# Basis fungsi
lnx = np.log(x)
cosx = np.cos(x)
ex = np.exp(x)

# ========================
# HITUNG SEMUA Σ (SEPERTI TABEL)
# ========================
S_ln2 = np.sum(lnx**2)
S_lncos = np.sum(lnx * cosx)
S_lnexp = np.sum(lnx * ex)

S_cos2 = np.sum(cosx**2)
S_cosexp = np.sum(cosx * ex)

S_exp2 = np.sum(ex**2)

S_yln = np.sum(y * lnx)
S_ycos = np.sum(y * cosx)
S_yexp = np.sum(y * ex)

# ========================
# BENTUK SISTEM PERSAMAAN
# ========================
A = np.array([
    [S_ln2,   S_lncos,  S_lnexp],
    [S_lncos, S_cos2,   S_cosexp],
    [S_lnexp, S_cosexp, S_exp2]
])

B = np.array([S_yln, S_ycos, S_yexp])

# ========================
# SOLVE
# ========================
a, b, c = np.linalg.solve(A, B)

print("Koefisien:")
print(f"a = {a}")
print(f"b = {b}")
print(f"c = {c}")

# ========================
# MODEL
# ========================
def f(x):
    return a*np.log(x) + b*np.cos(x) + c*np.exp(x)

# ========================
# PLOT
# ========================
x_plot = np.linspace(min(x), max(x), 200)
y_plot = f(x_plot)

plt.scatter(x, y, label="Data")
plt.plot(x_plot, y_plot, linestyle='--', label="Fit")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Example 4 - Least Squares Fit")
plt.legend()
plt.grid()

plt.show()
```

## Menggunakan Polynomial Orde 2

$$
f(x) = a_0 + a_1 x + a_2 x^2
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.array([0.24, 0.65, 0.95, 1.24, 1.73, 2.01, 2.23, 2.52])
y = np.array([0.23, -0.23, -1.1, -0.45, 0.27, 0.1, -0.29, 0.24])

# Polynomial orde 2 (quadratic)
coeffs = np.polyfit(x, y, 2)

# Ambil koefisien
a2, a1, a0 = coeffs
print("f(x) =", a2, "x^2 +", a1, "x +", a0)

# Fungsi polynomial
def f(x):
    return a2*x**2 + a1*x + a0

# Plot
x_plot = np.linspace(min(x), max(x), 200)
y_plot = f(x_plot)

plt.scatter(x, y, label="Data")
plt.plot(x_plot, y_plot, linestyle='--', label="Polynomial Fit (orde 2)")

plt.legend()
plt.grid()
plt.show()
```

## Menggunakan Polynomial Orde 3

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.array([0.24, 0.65, 0.95, 1.24, 1.73, 2.01, 2.23, 2.52])
y = np.array([0.23, -0.23, -1.1, -0.45, 0.27, 0.1, -0.29, 0.24])

# Polynomial orde 2 (quadratic)
coeffs = np.polyfit(x, y, 3)

# Ambil koefisien
a3, a2, a1, a0 = coeffs
print("f(x) =", a3, "x^3 +", a2, "x^2 +", a1, "x +", a0)

# Fungsi polynomial
def f(x):
    return a3*x**3 + a2*x**2 + a1*x + a0

# Plot
x_plot = np.linspace(min(x), max(x), 200)
y_plot = f(x_plot)

plt.scatter(x, y, label="Data")
plt.plot(x_plot, y_plot, linestyle='--', label="Polynomial Fit (orde 3)")

plt.legend()
plt.grid()
plt.show()
```
