# Penyelesaian Persamaan Nonlinear
Persamaan yang akan diselesaikan:

$$
f(x) = x - \cos(x) = 0
$$

Metode yang digunakan:
1. Newton-Raphson
2. Bisection
3. Secant

```python
import math

# Fungsi
def f(x):
    return x - math.cos(x)

# Turunan untuk Newton
def df(x):
    return 1 + math.sin(x)


# =========================
# 1️⃣ NEWTON METHOD
# =========================
def newton(x0, tol=1e-14, max_iter=100):
    x = x0
    for i in range(max_iter):
        x_new = x - f(x)/df(x)
        if abs(x_new - x) < tol:
            print(f"Newton converged in {i+1} iterations")
            return x_new
        x = x_new
    return x


# =========================
# 2️⃣ BISECTION METHOD
# =========================
def bisection(a, b, tol=1e-14, max_iter=100):
    if f(a)*f(b) > 0:
        print("Interval tidak valid")
        return None
    
    for i in range(max_iter):
        c = (a + b) / 2
        
        if abs(b - a) < tol:
            print(f"Bisection converged in {i+1} iterations")
            return c
        
        if f(a)*f(c) < 0:
            b = c
        else:
            a = c
            
    return (a + b) / 2


# =========================
# 3️⃣ SECANT METHOD
# =========================
def secant(x0, x1, tol=1e-14, max_iter=100):
    for i in range(max_iter):
        x2 = x1 - f(x1)*(x1 - x0)/(f(x1) - f(x0))
        
        if abs(x2 - x1) < tol:
            print(f"Secant converged in {i+1} iterations")
            return x2
        
        x0 = x1
        x1 = x2
        
    return x1


# =========================
# TESTING SESUAI SOAL
# =========================

print("Newton Method:")
root_newton = newton(0.8)
print("Root =", root_newton)

print("\nBisection Method:")
root_bisect = bisection(0.6, 0.8)
print("Root =", root_bisect)

print("\nSecant Method:")
root_secant = secant(0.6, 0.8)
print("Root =", root_secant)
```
