# Penyelesaian Persamaan Nonlinear
Persamaan yang akan diselesaikan:

$$
f(x) = x - \cos(x) = 0
$$

Metode yang digunakan:
1. Bisection
2. Newton
3. Secant

## Methode Bisection

```python
import numpy as np

#Fungsi soal:
def fungsi(x):
    return x-np.cos(x)

#mencari interval a dan b apakah bisa untuk bisection?
#syarat : jika kedua fungsi berbeda tandanya, maka bisa digunakan untuk bisection
print ("hasil untuk interval a: 0.6 = ", fungsi(0.6)) 
print ("hasil untuk interval b: 0.8 = ", fungsi(0.8))

if fungsi(0.6)*fungsi(0.8) < 0:
    print ("bisa digunakan untuk bisection")
else:
    print ("tidak bisa digunakan untuk bisection")
```

```python
def bisection(a,b,eps):
    u = fungsi(a) #interval a
    v = fungsi(b) #interval b
    i = 0
    erroriterasi = (b-a)/2 #error iterasi

    # Header tabel
    print(f"{'Iter':<5}{'a':<12}{'b':<12}{'c':<12}{'f(a)':<12}{'f(b)':<12}{'f(c)':<12}{'Error':<12}")
    print("-"*85)

    while (erroriterasi > eps): #akan mengulang iterasi selama error iterasi lebih besar dari eps
        i +=1
        erroriterasi = (b-a)/2 #error iterasi
        c = (a+b)/2 #mencari titik tengah
        w = fungsi(c) #fungsi titik tengah
        

        if w == 0: #jika titik tengah adalah akar, maka selesai
            print("akar ditemukan: ", c)
            return c
        elif u*w < 0: #jika fungsi a dan fungsi titik tengah berbeda tanda, maka akar berada di interval a dan titik tengah
            b = c #interval b diganti c
            v = w #fungsi b diganti fungsi c
        else: #jika fungsi b dan fungsi titik tengah berbeda tanda, maka akar berada di interval titik tengah dan b
            a = c #interval a diganti c
            u = w #fungsi a diganti fungsi c

        #print("iterasi","nilai a","nilai b","nilai c", "f(a)", "f(b)", "f(c)","error iterasi")
        print(f"{i:<5}{a:<12.6f}{b:<12.6f}{c:<12.6f}{fungsi(a):<12.6f}{fungsi(b):<12.6f}{fungsi(c):<12.6f}{erroriterasi:<12.6f}")

#bisection(0.6,0.8,0.00010)
bisection(0.6,0.8,1e-13) #kalau di ppt, diminta more than 13 correct digits, maka eps = 1e-13
```

## Metode Newton-Raphson

```python
import numpy as np

#Fungsi soal:
def fungsi(x):
    return x-np.cos(x)

#Turunan fungsi soal:
def turunan(x):
    return 1+np.sin(x)  #turunan dari x adalah 1
                        #turunan dari -cos(x) adalah sin(x)

def newton(x0,eps):
    i = 0
    erroriterasi = float('inf') #inisialisasi error iterasi dengan nilai tak hingga

    # Header tabel
    print(f"{'Iter':<5}{'x':<12}{'f(x)':<12}{'turunan(x)':<12}{'Error':<12}")
    print("-"*65)

    while (erroriterasi > eps): #akan mengulang iterasi selama error iterasi lebih besar dari eps
        i +=1
        fx = fungsi(x0) #fungsi pada titik x0
        dfx = turunan(x0) #turunan fungsi pada titik x0

        if dfx == 0: #jika turunan sama dengan nol, maka tidak bisa dilanjutkan karena akan terjadi pembagian dengan nol
            print("Turunan sama dengan nol. Tidak bisa dilanjutkan.")
            return None

        x1 = x0 - fx/dfx #rumus Newton-Raphson untuk mencari titik berikutnya
        erroriterasi = abs(x1 - x0) #error iterasi adalah selisih absolut antara titik baru dan titik lama

        #print("iterasi","nilai x","f(x)","f'(x)","error iterasi")
        print(f"{i:<5}{x1:<12.6f}{fx:<12.6f}{dfx:<12.6f}{erroriterasi:<12.6f}")

        x0 = x1 #update x0 untuk iterasi berikutnya

newton(0.8,1e-13) #kalau di ppt, diminta more than 13 correct digits, maka eps = 1e-13
```

## Metode Secant

```python
import numpy as np

#Fungsi soal:
def fungsi(x):
    return x-np.cos(x)

def secant(x0,x1,eps):
    i = 0
    erroriterasi = float('inf') #inisialisasi error iterasi dengan nilai tak hingga

    # Header tabel
    print(f"{'Iter':<5}{'x0':<12}{'x1':<12}{'f(x0)':<12}{'f(x1)':<12}{'Error':<12}")
    print("-"*70)

    while (erroriterasi > eps): #akan mengulang iterasi selama error iterasi lebih besar dari eps
        i +=1
        fx0 = fungsi(x0) #fungsi pada titik x0
        fx1 = fungsi(x1) #fungsi pada titik x1

        if fx1 - fx0 == 0: #jika selisih fungsi pada x1 dan x0 sama dengan nol, maka tidak bisa dilanjutkan karena akan terjadi pembagian dengan nol
            print("Selisih fungsi pada x1 dan x0 sama dengan nol. Tidak bisa dilanjutkan.")
            return None

        x2 = x1 - fx1*(x1 - x0)/(fx1 - fx0) #rumus secant untuk mencari titik berikutnya
        erroriterasi = abs(x2 - x1) #error iterasi adalah selisih absolut antara titik baru dan titik lama

        #print("iterasi","nilai x0","nilai x1","f(x0)","f(x1)","error iterasi")
        print(f"{i:<5}{x0:<12.6f}{x1:<12.6f}{fx0:<12.6f}{fx1:<12.6f}{erroriterasi:<12.6f}")

        x0, x1 = x1, x2 #update x0 dan x1 untuk iterasi berikutnya

secant(0.6,0.8,1e-13) #kalau di ppt, diminta more than 13 correct digits, maka eps = 1e-13
```
