# Linear Interpolation

## Linear Interpolation (Menentukan $\sin(0.15)$)


Diketahui dari tabel:

- $x_0 = 0.1,\quad f(x_0) = 0.0998$
- $x_1 = 0.2,\quad f(x_1) = 0.1987$

### Rumus Interpolasi Linear

$$
f(x) = f(x_0) + \frac{f(x_1) - f(x_0)}{x_1 - x_0}(x - x_0)
$$

### Perhitungan

Kemiringan (slope):

$$
m = \frac{0.1987 - 0.0998}{0.2 - 0.1}
= \frac{0.0989}{0.1}
= 0.989
$$

Nilai interpolasi:

$$
f(0.15) = 0.0998 + 0.989(0.15 - 0.1)
$$

$$
= 0.0998 + 0.989(0.05)
$$

$$
= 0.0998 + 0.04945
$$

$$
= 0.14925
$$

### Hasil Akhir

$$
f(0.15) \approx 0.1493
$$

```python
# Linear Interpolation untuk sin(0.15)

# data dari tabel
x0 = 0.1
y0 = 0.0998

x1 = 0.2
y1 = 0.1987

x = 0.15

# hitung slope
m = (y1 - y0) / (x1 - x0)

# hitung interpolasi
f_x = y0 + m * (x - x0)

# output
print("Slope (m) =", m)
print("f(0.15) =", f_x)
```

## Linear Interpolation (Viscosity pada $T = 8^\circ C$)


Diketahui dua titik:

- $(5, 1.519)$
- $(10, 1.308)$

### Rumus Interpolasi Linear

$$
V(T) = V(T_0) + \frac{V(T_1) - V(T_0)}{T_1 - T_0}(T - T_0)
$$

### Perhitungan

Kemiringan (slope):

$$
m = \frac{1.308 - 1.519}{10 - 5}
= \frac{-0.211}{5}
= -0.0422
$$

Nilai interpolasi:

$$
V(8) = 1.519 + (-0.0422)(8 - 5)
$$

$$
= 1.519 + (-0.0422)(3)
$$

$$
= 1.519 - 0.1266
$$

$$
= 1.3924
$$

### Hasil Akhir

$$
V(8) = 1.3924
$$

```python
# Linear Interpolation (Viscosity)

# data titik
T0 = 5
V0 = 1.519

T1 = 10
V1 = 1.308

T = 8

# hitung slope
m = (V1 - V0) / (T1 - T0)

# hitung interpolasi
V = V0 + m * (T - T0)

# output
print("Slope (m) =", m)
print("V(8) =", V)
```

# Divided Difference

## Example 3: Divided Difference

Diketahui:

- $x = [0, 1, -1]$
- $f(x) = [-5, -3, -15]$

### Rumus

$$
f[x_i, x_j] = \frac{f(x_j) - f(x_i)}{x_j - x_i}
$$

### Perhitungan

Orde 1:

$$
f[0,1] = \frac{-3 - (-5)}{1 - 0} = 2
$$

$$
f[1,-1] = \frac{-15 - (-3)}{-1 - 1} = 6
$$

Orde 2:

$$
f[0,1,-1] = \frac{6 - 2}{-1 - 0} = -4
$$

### Tabel

$$
\begin{bmatrix}
-5 & 2 & -4 \\
-3 & 6 & 0 \\
-15 & 0 & 0
\end{bmatrix}
$$

```python
x = [0, 1, -1]
y = [-5, -3, -15]

n = len(x)

# buat tabel kosong
dd = [[0 for _ in range(n)] for _ in range(n)]

# isi kolom pertama
for i in range(n):
    dd[i][0] = y[i]

# hitung divided difference
for j in range(1, n):
    for i in range(n - j):
        dd[i][j] = (dd[i+1][j-1] - dd[i][j-1]) / (x[i+j] - x[i])

# print tabel
for row in dd:
    print(row)
```

# Newton Interpolation Polynomial

## Example 4: Newton Interpolation Polynomial

Dari divided difference:

- $f[x_0] = -5$
- $f[x_0,x_1] = 2$
- $f[x_0,x_1,x_2] = -4$

### Rumus Newton

$$
f(x) = f[x_0] + f[x_0,x_1](x - x_0) + f[x_0,x_1,x_2](x - x_0)(x - x_1)
$$

### Substitusi

$$
f(x) = -5 + 2(x - 0) -4(x - 0)(x - 1)
$$

### Hasil

$$
f(x) = -4x^2 + 6x -5
$$

```python
# koefisien dari divided difference
f0 = -5
f01 = 2
f012 = -4

# tampilkan bentuk polynomial
print("f(x) = -5 + 2(x) -4(x)(x-1)")

# cek nilai di x tertentu
def f(x):
    return f0 + f01*(x - 0) + f012*(x - 0)*(x - 1)

print("f(2) =", f(2))
```

# Lagrange Interpolation

## Example 5: Lagrange Interpolation

Diketahui:

- $x = [\frac{1}{3}, \frac{1}{4}, 1]$
- $y = [2, -1, 7]$

### Rumus

$$
P(x) = \sum_{i=0}^{n} y_i L_i(x)
$$

$$
L_i(x) = \prod_{j \ne i} \frac{x - x_j}{x_i - x_j}
$$

```python
X = [1/3, 1/4, 1]
Y = [2, -1, 7]

def lagrange(x):
    result = 0
    n = len(X)

    for i in range(n):
        term = Y[i]
        for j in range(n):
            if i != j:
                term *= (x - X[j]) / (X[i] - X[j])
        result += term

    return result

print("P(0.5) =", lagrange(0.5))
```

# Interpolasi Banyak Titik

## Example 6: Interpolasi Banyak Titik

Diketahui:

$(0,1), (1,3), (2,2), (3,5), (4,4)$

Polinomial derajat 4 dapat dibentuk menggunakan metode Newton atau Lagrange.

```python
X = [0,1,2,3,4]
Y = [1,3,2,5,4]

def lagrange(x):
    result = 0
    n = len(X)

    for i in range(n):
        term = Y[i]
        for j in range(n):
            if i != j:
                term *= (x - X[j]) / (X[i] - X[j])
        result += term

    return result

print("P(2.5) =", lagrange(2.5))
```

# Newton Banyak Titik

```python
x = [2,4,5,6,7]
y = [3,5,1,6,9]

def divided_diff(x, y):
    n = len(x)
    table = [[0]*n for _ in range(n)]

    for i in range(n):
        table[i][0] = y[i]

    for j in range(1, n):
        for i in range(n - j):
            table[i][j] = (table[i+1][j-1] - table[i][j-1]) / (x[i+j] - x[i])

    return table

table = divided_diff(x, y)

for row in table:
    print(row)
```

# Inverse Interpolation

## Example 8: Inverse Interpolation

Diketahui:

- $x = [1,2,3]$
- $y = [3.2,2.0,1.6]$

Langkah:
1. Tukar $x$ dan $y$
2. Lakukan interpolasi

```python
X = [3.2, 2.0, 1.6]
Y = [1, 2, 3]

def lagrange(x):
    result = 0
    n = len(X)

    for i in range(n):
        term = Y[i]
        for j in range(n):
            if i != j:
                term *= (x - X[j]) / (X[i] - X[j])
        result += term

    return result

print("Hasil inverse =", lagrange(2.5))
```
