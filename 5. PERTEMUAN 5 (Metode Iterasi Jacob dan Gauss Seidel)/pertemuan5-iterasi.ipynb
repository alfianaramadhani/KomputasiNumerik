# Example 1 

Soal
$$
\begin{cases}
3x + y - z = 5 \\
4x + 7y - 3z = 20 \\
2x - 2y + 5z = 10
\end{cases}
$$

Bentuk Iterasi untuk x,y,z
$$
x = \frac{5 - y + z}{3}, \quad
y = \frac{20 - 4x + 3z}{7}, \quad
z = \frac{10 - 2x + 2y}{5}
$$

## Metode Iterasi Jacobian

```python
import pandas as pd

def jacobi_method(x0, y0, z0, tol=1e-6, max_iter=10):
    # inisialisasi
    x, y, z = x0, y0, z0
    
    # simpan hasil
    data = []
    
    print("Iter |    x        Ea_x(%) |    y        Ea_y(%) |    z        Ea_z(%)")
    print("-"*75)
    
    for i in range(1, max_iter+1):
        x_old, y_old, z_old = x, y, z
        
        # rumus Jacobi (pakai nilai lama semua)
        x_new = (5 - y_old + z_old) / 3
        y_new = (20 - 4*x_old + 3*z_old) / 7
        z_new = (10 - 2*x_old + 2*y_old) / 5
        
        # hitung error relatif (%)
        ea_x = abs((x_new - x_old)/x_new)*100 if x_new != 0 else 0
        ea_y = abs((y_new - y_old)/y_new)*100 if y_new != 0 else 0
        ea_z = abs((z_new - z_old)/z_new)*100 if z_new != 0 else 0
        
        # tampilkan
        print(f"{i:>4} | {x_new:>8.4f} {ea_x:>10.3f} | {y_new:>8.4f} {ea_y:>10.3f} | {z_new:>8.4f} {ea_z:>10.3f}")
        
        # simpan ke list
        data.append([i, x_new, ea_x, y_new, ea_y, z_new, ea_z])
        
        # cek konvergensi
        if ea_x < tol and ea_y < tol and ea_z < tol:
            print(f"\nKonvergen pada iterasi ke-{i}")
            break
        
        # update nilai
        x, y, z = x_new, y_new, z_new
    
    # buat dataframe
    df = pd.DataFrame(data, columns=[
        "Iterasi", "x", "Ea_x (%)",
        "y", "Ea_y (%)",
        "z", "Ea_z (%)"
    ])
    
    return df, (x, y, z)

# tebakan awal
x0, y0, z0 = 0, 0, 0

df, hasil = jacobi_method(x0, y0, z0, max_iter=10)

print("\nHasil akhir:")
print(f"x = {hasil[0]:.5f}, y = {hasil[1]:.5f}, z = {hasil[2]:.5f}")
```

## Metode Gauss-Seidel

```python
import pandas as pd

def gauss_seidel_method(x0, y0, z0, tol=1e-6, max_iter=10):
    # inisialisasi
    x, y, z = x0, y0, z0
    
    # simpan hasil
    data = []
    
    print("Iter |    x        Ea_x(%) |    y        Ea_y(%) |    z        Ea_z(%)")
    print("-"*75)
    
    for i in range(1, max_iter+1):
        x_old, y_old, z_old = x, y, z
        
        # Gauss-Seidel (pakai nilai terbaru langsung)
        x = (5 - y + z) / 3
        y = (20 - 4*x + 3*z) / 7
        z = (10 - 2*x + 2*y) / 5
        
        # hitung error relatif (%)
        ea_x = abs((x - x_old)/x)*100 if x != 0 else 0
        ea_y = abs((y - y_old)/y)*100 if y != 0 else 0
        ea_z = abs((z - z_old)/z)*100 if z != 0 else 0
        
        # tampilkan
        print(f"{i:>4} | {x:>8.4f} {ea_x:>10.3f} | {y:>8.4f} {ea_y:>10.3f} | {z:>8.4f} {ea_z:>10.3f}")
        
        # simpan ke list
        data.append([i, x, ea_x, y, ea_y, z, ea_z])
        
        # cek konvergensi
        if ea_x < tol and ea_y < tol and ea_z < tol:
            print(f"\nKonvergen pada iterasi ke-{i}")
            break
    
    # buat dataframe
    df = pd.DataFrame(data, columns=[
        "Iterasi", "x", "Ea_x (%)",
        "y", "Ea_y (%)",
        "z", "Ea_z (%)"
    ])
    
    return df, (x, y, z)


# tebakan awal
x0, y0, z0 = 0, 0, 0

df_gs, hasil_gs = gauss_seidel_method(x0, y0, z0, max_iter=10)

print("\nHasil akhir:")
print(f"x = {hasil_gs[0]:.5f}, y = {hasil_gs[1]:.5f}, z = {hasil_gs[2]:.5f}")
```

```python
import pandas as pd

def gauss_seidel_ex1(a1, a2, a3, max_iter=6):
    data = []
    
    print("Iter |   a1        Ea(%) |   a2        Ea(%) |   a3        Ea(%)")
    print("-"*75)
    
    for i in range(1, max_iter+1):
        a1_old, a2_old, a3_old = a1, a2, a3
        
        # update Gauss-Seidel
        a1 = (106.8 - 5*a2 - a3) / 25
        a2 = (177.2 - 64*a1 - a3) / 8
        a3 = 279.2 - 144*a1 - 12*a2
        
        # error relatif (%)
        ea1 = abs((a1 - a1_old)/a1)*100 if a1 != 0 else 0
        ea2 = abs((a2 - a2_old)/a2)*100 if a2 != 0 else 0
        ea3 = abs((a3 - a3_old)/a3)*100 if a3 != 0 else 0
        
        print(f"{i:>4} | {a1:>8.4f} {ea1:>10.3f} | {a2:>8.4f} {ea2:>10.3f} | {a3:>8.4f} {ea3:>10.3f}")
        
        data.append([i, a1, ea1, a2, ea2, a3, ea3])
    
    df = pd.DataFrame(data, columns=[
        "Iterasi","a1","Ea1 (%)","a2","Ea2 (%)","a3","Ea3 (%)"
    ])
    
    return df, (a1, a2, a3)

# initial guess dari slide
a1, a2, a3 = 1, 2, 5

df, hasil = gauss_seidel_ex1(a1, a2, a3, max_iter=6)

print("\nHasil akhir:")
print(f"a1 = {hasil[0]:.5f}, a2 = {hasil[1]:.5f}, a3 = {hasil[2]:.5f}") 
print("Hasil tidak konvergen karena errornya malah semakin membesar, karena matriks tidak diagonal dominan")
```
